## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
#讀取資料
import requests
data=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=data.text
response
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import json 
data=json.loads(response)
print('一次會回傳',len(data),'筆資料')
print('---')
print('每筆資料包含欄位有',data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

import json
data=json.loads(response)
for js in data:
    print('標題',js['title'])
    print('貼文時間',js['createdAt'])
    print('留言人數',js['commentCount'])
    print('按讚人數',js['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

comment_sum=0
like_sum=0

for d in data:
    comment_sum+=d['commentCount']
    sum=comment_sum
    
print('熱門文章')
print('平均留言人數:', sum/len(data))

sum=0
for d in data:
    like_sum+=d['likeCount']
    sum=like_sum
print('平均按讚人數:', sum/len(data))
print("")

comment_sum=0
like_sum=0
sum=0

for d in data:
    comment_sum+=d['commentCount']
    sum=comment_sum
    
print('非熱門文章')
print('平均留言人數:', sum/len(data))

sum=0
for d in data:
     like_sum+=d['likeCount']
     sum=like_sum
print('平均按讚人數:', sum/len(data))
```
